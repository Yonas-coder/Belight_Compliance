```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
from sklearn.impute import SimpleImputer


```

```python
### Datasets: 
train = pd.read_csv('train.csv',encoding='ISO-8859-1',low_memory=False) #encoding='cp1252')  
test = pd.read_csv('test.csv',encoding='cp1252',low_memory=False)

address = pd.read_csv('addresses.csv')
latlons = pd.read_csv('latlons.csv')
```

```python
### Merge datasets: mapping from ticket_id to addresses, and from addresses to lat/lon coordinates

geolocated = pd.merge(address,latlons, on = 'address')
print(geolocated.columns)
geolocated.drop('address', axis=1,inplace=True)

train= pd.merge(train,geolocated, on='ticket_id')
test= pd.merge(test,geolocated, on='ticket_id')
del address, latlons, geolocated

print(train.shape)
print(test.shape)
train.compliance.value_counts()
```

```python
#Drop the violation addresses in train and test datasets that are already mapped by the lon/lat coordinates: 
train.drop(['violation_street_number', 'violation_zip_code'],
           axis=1, inplace=True)
test.drop(['violation_street_number', 'violation_zip_code'],
           axis=1, inplace=True)

###############################

# Remove columns that are only in train dataset but not available during testing. 
train.drop(['payment_amount','payment_date', 'payment_status', 'collection_status', 
            'balance_due', 'compliance_detail'], axis=1, inplace=True)

#train = train.iloc[list(test.index)]

##############################

# Setting the ticket_id as the index
train.set_index('ticket_id',inplace=True)
test.set_index('ticket_id',inplace=True)

##############################
```

### Handling missing values:

```python
#Print Missing values:

nan_train = train.isnull().sum()

print('Total missing values per feature in the train dataset:')
df = pd.DataFrame(data={'Total_Missing':nan_train[nan_train>0],'Data Type':train.dtypes[nan_train[nan_train>0].index]})
print(df.sort_values('Total_Missing'))

print('\n')

nan_test = test.isnull().sum()

print('Total missing values per feature in the test dataset:')
df = pd.DataFrame(data={'Total_Missing':nan_test[nan_test>0],'Data Type':test.dtypes[nan_test[nan_test>0].index]})
print(df.sort_values('Total_Missing'))

test['grafitti_status'].value_counts()
```

```python
# Remove the rows with Null value for compliance(target variable for prediction), 
# those are found not responsible and we don't use the entries for training.
train= train[np.isfinite(train['compliance'])]
```

```python
# For the lat and lon missing values, linear interpolation is used to replace nan values
train.lat.interpolate(method='linear', inplace=True) 
train.lon.interpolate(method='linear', inplace=True) 
test.lat.interpolate(method='linear', inplace=True) 
test.lon.interpolate(method='linear', inplace=True) 

#there is one sample with missing fine_amount value in train, I use the most frequent fine amount to fill it 
imp = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
train['fine_amount'] = list(imp.fit_transform(np.array(list(train['fine_amount'])).reshape(-1, 1)).flatten())
test['fine_amount'] =  list(imp.transform(np.array(list(test['fine_amount'])).reshape(-1, 1)).flatten())   # no missing for the current test data

# There is only one entry(=GRAFFITI TICKET) for the grafitti_status in the train dataset. 
# There are 2221 entires(=GRAFFITI TICKET) for the grafitti_status in the test dataset. 
# Instead of dropping this column, I will impute 'missing' for the missing values 
train['grafitti_status'].fillna(value='missing', inplace=True)  
test['grafitti_status'].fillna(value='missing', inplace=True)  
```

###############################

Every sample has different mailing address values, Therefore, I grouped mailing addresses into four parts: 
- P.O. Box  - if entry for 'mailing_address_str_name' is P.O. Box
- Same, - if it's the same as where the violation occured 
   - (wherever there is no entry for mailing_address_str_name)
- Different, - if it's different from the address where the violation occured 
   - wherever there is an entry other than P.O. Box, it's different from 'violation_street_name'
   - any([train['MailingAddress'][i]==train['violation_street_name'][i] for i in list(train.index)])  #=False
- Foreign  - if mailing address is outside USA, indicated in 'country'
 
I create a column indicating these representation values and remove the following mailing-address related columns
 'mailing_address_str_number', 'mailing_address_str_name', 'city', 'state', 'zip_code', 'country' 
 Here, the data contains the phrase P.O. Box in several ways, i.e. POBOX, P.O.Box, P. O. Box, P.O. box, etc. I
 used regex to find all of them 

```python
print(train['country'].unique()) 
```

```python
def mailAddress_grouper(df_temp):
    df_temp['MailingAddress'] = 0 #'Different'
    POB_ind = df_temp['mailing_address_str_name'].str.contains(r'\b[Pp]+[.\s]*[Oo]+[.\s]*[Bb]{1,1}', na=False)
    df_temp.loc[POB_ind,'MailingAddress'] = 1 #'P.O.BOX'

    same_ind = df_temp['mailing_address_str_name'].isnull()
    df_temp.loc[same_ind,'MailingAddress'] = 2 #'Same'
    same_ind2 = list(df_temp['mailing_address_str_name'].str.lower()==df_temp['violation_street_name'].str.lower())
    df_temp.loc[same_ind2,'MailingAddress'] = 2 #'Same'

    Foreign_ind = df_temp['country']!= 'USA'
    df_temp.loc[same_ind,'MailingAddress'] = 3 #'Foreign'

    
    return df_temp['MailingAddress']

train['MailingAddress'] = mailAddress_grouper(train[['mailing_address_str_name','violation_street_name','country']].copy())
train.drop(['mailing_address_str_number', 'mailing_address_str_name', 'city', 'state', 'zip_code', 
                'non_us_str_code','country','violation_street_name'], axis=1, inplace=True)
#for Test dataset:

test['MailingAddress'] = mailAddress_grouper(test[['mailing_address_str_name','violation_street_name','country']].copy())
test.drop(['mailing_address_str_number', 'mailing_address_str_name', 'city', 'state', 'zip_code', 
                'non_us_str_code','country','violation_street_name'], axis=1, inplace=True)
```

```python
# 'violator_name': 
# this column is dropped because many people in the city may have the same name, 
# and learning on names may introduce some bias, two people with the same name should be considered different

train.drop(['violator_name'],axis=1, inplace=True)
test.drop(['violator_name'],axis=1, inplace=True)

# 'inspector_name': 
# there should be limited number of people who work for the city as inspectors, 
# their employee_id number is not provided but I can get the list of all inspector names in the test and 
# train datasets and incode their name with numerical values.  

inspector_in_train = train['inspector_name'].str.lower().unique()
print('Number of inpspectors in train dataset: ',len(inspector_in_train))
inspector_in_test = test['inspector_name'].str.lower().unique()
print('Number of inpspectors in test dataset: ',len(inspector_in_test))
All_instpector_names = list(np.union1d(inspector_in_test,inspector_in_train))
print('Union set of inspectors from the two lists: ',len(All_instpector_names))
_id = list(np.arange(len(All_instpector_names)))
inspector_id = pd.Series(_id, index = All_instpector_names)

train['inspector_id'] = train['inspector_name'].str.lower().apply(lambda x: inspector_id[x] )
test['inspector_id'] = test['inspector_name'].str.lower().apply(lambda x: inspector_id[x] )

train.drop('inspector_name',axis=1,inplace=True)
test.drop('inspector_name',axis=1,inplace=True)
```

```python
##############################
# The target variable is 'compliance', which is 'True' if the ticket was paid early, on time, or 
# within one month of the hearing date, 'False' if the ticket was paid after the hearing date or not at all.
# Therefore, creating a feature for number of days between ticket_issued_date and hearing_date is important. 
# and the model should be trained on this feature instead of the actual dates, which are important for future data predictions. 

#Get the time difference between hearing_date and ticket_issued_date
hd = pd.to_datetime(train.hearing_date,errors='coerce')
td = pd.to_datetime(train.ticket_issued_date,errors='coerce')
train['DayGaps'] = (hd - td).dt.days
train['DayGaps'].fillna(value=0, inplace=True)
train['DayGaps'] = train['DayGaps'].astype(int)

hd_test = pd.to_datetime(test.hearing_date,errors='coerce')
td_test = pd.to_datetime(test.ticket_issued_date,errors='coerce')
test['DayGaps'] = (hd_test - td_test).dt.days
test['DayGaps'].fillna(value=0, inplace=True)
test['DayGaps'] = test['DayGaps'].astype(int)

train.drop(['hearing_date','ticket_issued_date'],axis=1, inplace=True)
test.drop(['hearing_date','ticket_issued_date'],axis=1, inplace=True)
```
